```python
#ADD append rule for files

def inspect_attributes(obj):
    instance_data = vars(obj)
    content = []
    for name, value in instance_data.items():
        if not name.startswith('_') and value is not None and (not isinstance(value, list) or len(value) > 0):
            if isinstance(value, list):
                list_str = ", ".join(str(item) for item in value)
                content.append(f"{name}= {list_str}")
            else: content.append(f"{name}= {value}")
    return content

contents = inspect_attributes(file)
print(contents)
```

```python
with open(f"{file.name}.inp", "w", encoding="utf-8") as f:
    f.writelines('\n'.join(contents))
    print(contents)
```
