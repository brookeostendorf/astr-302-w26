```python
input_ = comp_files

create_file_button = widgets.Button(
    description='Create Input File',
    button_style='success'  
)

output = widgets.Output()

def on_button_clicked(b):
    with output:
        output.clear_output()
        contents = AttributeStringArray.inspect_attributes(input_)
        FileWriter.write_inp_file(contents)
        print("File created successfully.")

create_file_button.on_click(on_button_clicked)

display(create_file_button, output)
```
